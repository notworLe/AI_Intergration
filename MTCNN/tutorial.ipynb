# Usage example

```python
import cv2
import torch
from mtcnn import MTCNN
from mtcnn.utils.images import load_image

# create a detector instance
detector = MTCNN(device="GPU:0")

# load an image
# image = load_image('dataset/do_mi_xi/do1.jpg')
image = load_image('dataset/many_faces.jpg')

# detect faces in the image
result = detector.detect_faces(image)

print(result)
```

# MTCNN in pytorch

```python
from facenet_pytorch import MTCNN
import torch
import cv2
from myFunctions.notworlecv2 import read_display_img

device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
print(device)

# Threshold for P, R, O layers
mtcnn = MTCNN(thresholds=[0.7, 0.7, 0.8], keep_all=True, device=device)

image = cv2.imread('dataset/no_face.jpg')

box, probs = mtcnn.detect(image)
if box is not None:
    for face, prob in zip(box, probs):
        x1, y1, x2, y2 = [int(x) for x in face]
        cv2.rectangle(image, pt1=(x1, y1), pt2=(x2, y2), color=(173, 250, 93), thickness=2)

read_display_img(image)
```
