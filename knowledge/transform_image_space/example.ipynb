```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from pathlib import Path
from notworLeUtils.notworle_utils.matplotlib import draw_plot2


root = Path.cwd()
```

```python
reze = cv2.imread(root / ".." / "data" / "img" / "reze.jpg")
eris = cv2.imread(root / ".." / "data" / "img" / "eris.jpeg")

fig, axes = plt.subplots(1, 2, figsize=(10, 10))

images = [reze, eris]
titles = ['Reze', 'Eris']
draw_plot2(axes, titles, images, axis='on')
```

# Resize

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 10))

reze_resize = cv2.resize(reze, None, fx=1.5, fy=1.5)
eris_resize = cv2.resize(eris, None, fx=0.5, fy=0.5)

images = [reze_resize, eris_resize]
titles = ['Reze', 'Eris']

draw_plot2(axes, titles, images, axis='on')
```

# Rotate

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 10))

reze_rotate= cv2.rotate(reze, cv2.ROTATE_90_CLOCKWISE)
eris_rotate = cv2.rotate(eris, cv2.ROTATE_90_CLOCKWISE)

images = [reze_rotate, eris_rotate]
titles = ['Reze', 'Eris']

draw_plot2(axes, titles, images, axis='on')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 10))

h, w = reze.shape[:2]
reze_center = (w // 2, h // 2)
reze_rotate2d = cv2.getRotationMatrix2D(reze_center, angle=45, scale=1.0)
reze_rotate2d = cv2.warpAffine(reze, reze_rotate2d, (w, h))


h, w = eris.shape[:2]
print(w, h)
eris_center = (w // 2, h // 2)
eris_rotate2d = cv2.getRotationMatrix2D(eris_center, angle=120, scale=1.0)
eris_rotate2d = cv2.warpAffine(eris, eris_rotate2d, (w, h))

images = [reze_rotate2d, eris_rotate2d]
titles = ['Reze', 'Eris']

draw_plot2(axes, titles, images, axis='on')
```

# Translation

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 10))

h, w = reze.shape[:2]
tx, ty = 100, 50 # Dịch phải 100, dịch xuống 50
m = np.float32([
    [1, 0, tx],
    [0, 1, ty]
])
reze_translation = cv2.warpAffine(reze, m, (w, h))


h, w = eris.shape[:2]
tx, ty = 100, -500
m = np.float32([
    [1, 0, tx],
    [0, 1, ty]
])
eris_translation = cv2.warpAffine(eris, m, (w, h))

images = [reze_translation, eris_translation]
titles = ['Reze', 'Eris']

draw_plot2(axes, titles, images, axis='on')
```
