```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from pathlib import Path
from notworLeUtils.notworle_utils.matplotlib import draw_plot2


root = Path.cwd()
```

```python
reze = cv2.imread(root / ".." / "data" / "img" / "reze.jpg", cv2.IMREAD_GRAYSCALE)
eris = cv2.imread(root / ".." / "data" / "img" / "eris.jpeg", cv2.IMREAD_GRAYSCALE)
kobeni = cv2.imread(root / ".." / "data" / "img" / "kobeni.jpg", cv2.IMREAD_GRAYSCALE)

fig, axes = plt.subplots(1, 3, figsize=(10, 10))

images = [reze, eris, kobeni]
titles = ['Reze', 'Eris', 'Kobeni']
draw_plot2(axes, titles, images, axis='on')
```

# Create structure element

```python
kernel_rect  = cv2.getStructuringElement(cv2.MORPH_RECT,    (5, 5))
kernel_ellip = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
kernel_cross = cv2.getStructuringElement(cv2.MORPH_CROSS,   (5, 5))

print("RECT kernel:\n",    kernel_rect)
print("ELLIPSE kernel:\n", kernel_ellip)
print("CROSS kernel:\n",   kernel_cross)
```

# Erosion

```python
fix, axes = plt.subplots(1, 3, figsize=(10, 10))

# reze_erosion = cv2.erode(reze, kernel_rect, iterations=1)
reze_erosion = cv2.morphologyEx(reze, cv2.MORPH_ERODE, kernel_rect)

reze_erosion_x2 = cv2.erode(reze_erosion, kernel_rect, iterations=1)


titles = ["Reze" , "Erosion", "Erosion x2"]
images = [reze, reze_erosion, reze_erosion_x2]

draw_plot2(axes, titles, images)
```

# Dilation

```python
fix, axes = plt.subplots(1, 3, figsize=(10, 10))

# reze_erosion = cv2.dilate(reze, kernel_rect, iterations=1)
reze_dilate = cv2.morphologyEx(reze, cv2.MORPH_DILATE, kernel_rect)

reze_dilate_x2 = cv2.dilate(reze_dilate, kernel_rect, iterations=1)


titles = ["Reze" , "Dilate", "Dilate x2"]
images = [reze, reze_dilate, reze_dilate_x2]

draw_plot2(axes, titles, images)
```

# Opening

```python
fix, axes = plt.subplots(1, 3, figsize=(10, 10))

reze_open = cv2.morphologyEx(reze, cv2.MORPH_OPEN, kernel_rect)

reze_open_x2 = cv2.morphologyEx(reze_open, cv2.MORPH_OPEN, kernel_rect)


titles = ["Reze" , "Opening", "Opening x2"]
images = [reze, reze_open, reze_open_x2]

draw_plot2(axes, titles, images)
```
